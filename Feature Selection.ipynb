```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset=pd.read_csv('X_train.csv')
```

```python

dataset.head()
```

```python
y_train=dataset[['SalePrice']]
```

```python
X_train=dataset.drop(['Id','SalePrice'],axis=1)
```

```python
sel_ = SelectFromModel(Lasso(alpha=0.0001, random_state=0))
sel_.fit(X_train, y_train)

selected_features = X_train.columns[sel_.get_support()]

# total features
print('total features: {}'.format(X_train.shape[1]))
print('selected features: {}'.format(len(selected_features)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(sel_.estimator_.coef_ == 0)))
```

```python
sel_.get_support()
```

```python
for alpha in [0.005, 0.001, 0.0005, 0.0001]:
    model = SelectFromModel(Lasso(alpha=alpha, random_state=0))
    model.fit(X_train, y_train)
    n = model.get_support().sum()
    print(f"alpha={alpha} → {n} features selected")
```

```python
selected_feat = X_train.columns[sel_.get_support()]
selected_feat
```

```python
for alpha in [0.00005, 0.00001, 0.000005, 0.000001]:
    model = SelectFromModel(Lasso(alpha=alpha, random_state=0))
    model.fit(X_train, y_train)
    n = model.get_support().sum()
    print(f"alpha={alpha} → {n} features selected")
```

```python
X_train=X_train[selected_feat]
```

```python
# Finalize with alpha=0.0001
feature_sel_model = SelectFromModel(Lasso(alpha=0.0001, random_state=0))
feature_sel_model.fit(X_train, y_train)

# See which features were selected
selected_features = X_train.columns[feature_sel_model.get_support()]
print("Selected features:", selected_features.tolist())
print("Total:", len(selected_features))

# Filter your dataset
X_train_selected = feature_sel_model.transform(X_train)
print("New X_train shape:", X_train_selected.shape)
```
